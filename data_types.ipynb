# Python Data Types: definition + usage

```python


# Numeric
an_int =42
a_float = 3.14
a_complex = 1 + 2j
```

```python
an_int, a_float, a_complex
```

```python
#boolean
a_bool = True
```

```python
#String
a_string = "Hello, World!"
```

```python
#sequence types
a_list = [1, 2, 3, 4, 5]
a_tuple = (10, 20, 30)
a_range = range(0, 10)
```

```python
a_list, a_tuple, a_range
```

```python
#set types
a_set = {1, 2, 3, 4, 5} #unordered, mutable, no duplicates
a_frozenset = frozenset([10, 20, 30]) #unordered, immutable, no duplicates
```

```python
#Binary types
a_bytes = b'Hello' #immutable sequence of bytes
a_bytearray = bytearray(b'Hello') #mutable sequence of bytes
a_memoryview = memoryview(b'Hello') #memory view of bytes
```

```python
a_bytes, a_bytearray, a_memoryview
```

```python
# None type
nothing = None

values = [
    an_int, a_float, a_complex, a_bool, a_string,
    a_list, a_tuple, a_range,
    a_set, a_frozenset,
    a_bytes, a_bytearray, a_memoryview,
    nothing
]

for value in values:
    print(f"{value} is of type {type(value)}")
```

```python
x = "123"
x_int = int(x)
x_float = float(x)

num = [1,2,3,4,5]
num_tuple = tuple(num)

print(f"x: {x} (type: {type(x)})")
print(f"x_int: {x_int} (type: {type(x_int)})")
print(f"x_float: {x_float} (type: {type(x_float)})")
print(f"num: {num} (type: {type(num)})")
print(f"num_tuple: {num_tuple} (type: {type(num_tuple)})")
```
